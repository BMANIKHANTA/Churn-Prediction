# Handling Outliers Using Standradization

```python
def remove_outliers(df):
    mean = df.mean()
    std = df.std()
    lower_bounds = mean - 3 * std
    upper_bounds = mean + 3 * std
    # Iterate through each column in the DataFrame and filter out rows that fall within the bounds
    for col in df.columns:
        df = df[(df[col] > lower_bounds[col]) & (df[col] < upper_bounds[col])]
    # Print summary statistics for mean, standard deviation, lower bounds, and upper bounds    
    print("Means:",mean)
    print("\nSTD:",std)
    print("\nLower Bounds:",lower_bounds)
    print("\nUpper Bounds:",upper_bounds)
    # Return the DataFrame with outliers removed
    return df
```

```python
print(remove_outliers(df))
```

```python
df=remove_outliers(df)
```

# Missing Value Treatment

```python
# Display the initial number of missing values
print("Initial missing values:\n", df.isnull().sum())
```

```python
# 1. Remove records if NA's are less than 5%
threshold_5_percent = 0.05 * len(df)
df_cleaned = df.dropna(thresh=threshold_5_percent)
```

```python
df_cleaned
```

```python
# 2. Remove columns if NA's are 50% in any variable
threshold_50_percent = 0.5 * len(df)
df_cleaned = df_cleaned.dropna(axis=1, thresh=threshold_50_percent)
```

```python
df_cleaned
```

```python
# 3. Impute missing values
for column in df_cleaned.columns:
    if df_cleaned[column].dtype == 'object':
        # Categorical column: Impute with mode
        df_cleaned[column].fillna(df_cleaned[column].mode()[0], inplace=True)
    else:
        # Numerical column: Impute with mean
        df_cleaned[column].fillna(df_cleaned[column].mean(), inplace=True)
```

```python
df_cleaned
```

```python
print("Missing values after treatment:\n", df_cleaned.isnull().sum())
```

# Handling Highly Correlated Variables

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
# Generate a heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Initialize an empty list to store highly correlated variable pairs
highly_correlated_vars = []
# Iterate over the columns of the correlation matrix
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.9:
            highly_correlated_vars.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

print("Highly Correlated Variables:",highly_correlated_vars)
```

```python
# Remove one variable from each pair of highly correlated variables
vars_to_remove = []
for pair in highly_correlated_vars:
    # Remove the variable with lower importance 
    # For simplicity, let's remove the second variable in each pair
    vars_to_remove.append(pair[1])

print("Variables to Remove:",vars_to_remove)
```

```python
# Remove the variables
df.drop(columns=vars_to_remove, inplace=True)

print("Updated DataFrame Shape:", df.shape)
```

# Multicollinearity (VIF>5)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Separate target column (adjust the name 'target' as per your dataset)
target_column = 'target'
```

```python
# calculate VIF for each column
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif_data["feature"] = df.columns
    return vif_data
```

```python
print(vif_data)
```

```python
# Calculate VIF for all features
vif_data = calculate_vif(df)
```

```python
# Print columns with VIF < 5
high_vif_cols = vif_data[vif_data["VIF"] < 10]["feature"].tolist()
```

```python
print("Columns with VIF < 5:")
print(high_vif_cols)
```

```python
# Drop columns with VIF > 5
df.drop(columns=high_vif_cols, inplace=True)
```

```python
# Add the target column
df.append(target_column)
```

```python
print("Updated DataFrame after dropping columns with VIF > 5:")
print(df.shape)
```
